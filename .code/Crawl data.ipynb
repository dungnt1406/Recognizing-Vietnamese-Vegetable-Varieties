#2. CODE FOR CHECK NUMBER OF IMAGE IN EVERY FOLDER DATA

```python
import os
import pandas as pd

# Danh sách các thư mục cần sắp xếp
folders = ['0. Good Vietnamese Balm', '1. Average Vietnamese Balm',
           '2. Bad Vietnamese Balm', '3. Good Piper Lolot', '4. Average Piper Lolot',
           '5. Bad Piper Lolot', '6. Good Skunk Vine', '7. Average Skunk Vine',
           '8. Bad Skunk Vine', '9. Good Water Morning Glory', '10. Average Water Morning Glory',
           '11. Bad Water Morning Glory', '12. Good Perilla', '13. Average Perilla',
           '14. Bad Perilla', '15. Good Katuk', '16. Average Katuk', '17. Bad Katuk']

def count_images_in_subfolders(directory):
    """
    Đếm số lượng ảnh trong mỗi thư mục con của thư mục chính.

    Parameters:
    - directory (str): Đường dẫn đến thư mục chính.

    Returns:
    - pandas.DataFrame: DataFrame chứa số lượng ảnh trong mỗi thư mục con.
    """
    data = []
    for folder in folders:
        folder_path = os.path.join(directory, folder)
        for subfolder in ['Crawled online', 'Phone cameras']:
            subfolder_path = os.path.join(folder_path, subfolder)
            if os.path.exists(subfolder_path):
                num_images = sum([len(files) for r, d, files in os.walk(subfolder_path) if any(f.lower().endswith(('.jpg', '.jpeg', '.png')) for f in files)])
                data.append([folder, subfolder, num_images])

    df = pd.DataFrame(data, columns=['Folder', 'Subfolder', 'Number of Images'])
    return df

# Đường dẫn đến thư mục RAW DATA
directory_path = '/content/drive/MyDrive/DAP391m_G06_AI1807/Data/RAW DATA'

# Đếm số lượng ảnh trong mỗi thư mục con
df = count_images_in_subfolders(directory_path)

# Hiển thị DataFrame
df
```

#CODE FOR RENAME IMAGE IN FOLDER

```python
import os

# Định nghĩa các từ viết tắt
abbreviations = {
    'Crawled online': 'c',
    'Phone cameras': 'p',
    'Good': 'g',
    'Average': 'a',
    'Bad': 'b',
    'Vietnamese Balm': 'VNB',
    'Piper Lolot': 'PPL',
    'Skunk Vine': 'SKV',
    'Water Morning Glory': 'WMG',
    'Katuk': 'KTU',
    'Perilla': 'PER'
}

# Map các folder chính đến từ viết tắt
main_folders = {
    '0. Good Vietnamese Balm': 'VNB',
    '1. Average Vietnamese Balm': 'VNB',
    '2. Bad Vietnamese Balm': 'VNB',
    '3. Good Piper Lolot': 'PPL',
    '4. Average Piper Lolot': 'PPL',
    '5. Bad Piper Lolot': 'PPL',
    '6. Good Skunk Vine': 'SKV',
    '7. Average Skunk Vine': 'SKV',
    '8. Bad Skunk Vine': 'SKV',
    '9. Good Water Morning Glory': 'WMG',
    '10. Average Water Morning Glory': 'WMG',
    '11. Bad Water Morning Glory': 'WMG',
    '12. Good Perilla': 'PER',
    '13. Average Perilla': 'PER',
    '14. Bad Perilla': 'PER',
    '15. Good Katuk': 'KTU',
    '16. Average Katuk': 'KTU',
    '17. Bad Katuk': 'KTU'
}

def rename_images_in_subfolders(root_folder):
    for main_folder, main_abbr in main_folders.items():
        main_folder_path = os.path.join(root_folder, main_folder)
        quality_abbr = abbreviations[main_folder.split('.')[1].strip().split(' ')[0]]

        for subfolder, sub_abbr in {'Crawled online': 'c', 'Phone cameras': 'p'}.items():
            subfolder_path = os.path.join(main_folder_path, subfolder)
            if os.path.exists(subfolder_path):
                images = [f for f in os.listdir(subfolder_path) if f.lower().endswith(('.jpg', '.jpeg', '.png'))]

                for i, image in enumerate(images):
                    new_name = f"{quality_abbr}_{sub_abbr}_{main_abbr}_{i}.jpg"
                    old_image_path = os.path.join(subfolder_path, image)
                    new_image_path = os.path.join(subfolder_path, new_name)
                    os.rename(old_image_path, new_image_path)
                    print(f"Renamed {old_image_path} to {new_image_path}")

# Đường dẫn đến thư mục RAW DATA
directory_path = '/content/drive/MyDrive/DAP391m_G06_AI1807/Data/RAW DATA'

# Đổi tên các ảnh trong tất cả các subfolder
rename_images_in_subfolders(directory_path)
```
